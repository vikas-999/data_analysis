```python
#reading the csv file in pandas
import pandas as pd
df = pd.read_csv('ZillowApartmens_msp.csv')
df
```

```python
#information about the data frame
df.info()
```

```python
#we have price ranges in apt_price column so splitting those columns
df[['apt_price','apt_price2']] = df['apt_price'].str.split('-',1, expand = True)
```

```python
#after splitting
df
```

```python
#cleaning the data
df['apt_price'] = df['apt_price'].str.replace('$', '', regex=True)
df['apt_price'] = df['apt_price'].str.replace(',', '', regex=True)
df['apt_price2'] = df['apt_price2'].str.replace(',', '', regex=True)
df['apt_price'] = df['apt_price'].str.replace(' /mo', '', regex=True)
df['apt_price'] = df['apt_price'].str.replace('nan', '', regex=True)
df['apt_type'] = df['apt_type'].str.replace('NaN', '', regex=True)
df['apt_price'] = df['apt_price'].str.replace('NA', '', regex=True)
df
```

```python
#replacing the values so that to perfome operations on that column
df1 = df.replace({'apt_price': {'Call for Rent': ''}}, regex=True)
df1
```

```python
# convert string to an integer
df1['apt_price'] = pd.to_numeric(df1['apt_price'])
df1['apt_price2'] = pd.to_numeric(df1['apt_price2'])
df1.info()
```

```python
# fetching the properties lessesthan 1000$
bel_1000 = df1[df1['apt_price'] < 1000]
bel_1000
```

```python
# fetching the properties greaterthan 1500$
above_1500 = df1[df1['apt_price'] > 1500]
above_1500
```

```python
#fetching row in between range
df1.loc[df1['apt_price'].between(1500, 2000)]
```
